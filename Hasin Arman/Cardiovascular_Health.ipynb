```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Age': [45, 55, 60, 50, 65],
    'Cholesterol': [180, 200, 220, 190, 240],
    'BMI': [25, 30, 28, 26, 32],
    'SmokingHistory': ['Non-Smoker', 'Smoker', 'Non-Smoker', 'Smoker', 'Non-Smoker'],
    'HeartDisease': ['No', 'Yes', 'No', 'No', 'Yes']
}

heart_data = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
sns.histplot(heart_data, x='Age', hue='HeartDisease', multiple='stack', bins=10, kde=True, palette='Set1')
plt.title('Distribution of Age by Heart Disease Status')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(x='SmokingHistory', y='Cholesterol', data=heart_data, palette='coolwarm')
plt.title('Cholesterol Levels by Smoking History')
plt.xlabel('Smoking History')
plt.ylabel('Cholesterol Level')
plt.show()

sns.pairplot(heart_data, hue='HeartDisease', palette='husl')
plt.suptitle('Pairplot of Cardiovascular Health Factors', y=1.02)
plt.show()

plt.figure(figsize=(10, 8))
numeric_columns = heart_data.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_columns.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=1)
plt.title('Correlation Heatmap of Cardiovascular Health Factors')
plt.show()
```
